```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
orders_df = pd.read_csv("../American_Stone_Project/filetoanalyze.csv")
orders_df.head()
```

```python
x_axis = np.arange(len(orders_df))
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(20,3))
plt.bar(x_axis, orders_df["price"], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, orders_df["order_item_id"], rotation="vertical")
```
